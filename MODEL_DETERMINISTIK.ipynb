```python
# GERAK JATUH BEBAS
# INISASI
import numpy as np
import matplotlib.pyplot as plt
import math

# Input
y = 200
v = 0
a = -9.8
t = 0
dt = 0.2

# initialize array
y_arr = [y]
v_arr = [v]
t_arr = [t]
```

```python
#NUMERICAL
while y >= 0:
  v += a * dt
  y += v * dt
  t += dt
  if y < 0 :
    break

  v_arr.append(v)
  y_arr.append(y)
  t_arr.append(t)

print("vt (Numerical) :", y)
print("vt (Numerical) :", v)
print("t (Numerical) :", t_arr[-1])
```

```python
# ANALYTICAL
# exact solution ⟹ y(t) = y(0) + (1/2gt^2)
# v(t) = v(0) + g*t
v_ex = []
y_arr_ex = []
for t in t_arr:
  y_ex = y_arr[0] + (0.5 * a * t**2)
  y_arr_ex.append(y_ex)
  v_ex.append(a*t)
# total time ⟹ t_tot = sqrt((2*y(0)) / g)
t_ex = np.sqrt((2 * -y_arr[0])/a)


print("Analytical y(t) :", y_ex)
print("Analytical v(t) :", vt)
```

```python
print('Numerical vs Analytical')
print('yt (m): {:.2f} vs {:.2f}'.format(y, y_ex))
print('vt (m/s): {:.2f} vs {:.2f}'.format(v, vt))
```

```python
# plot output
plt.plot(t_arr, y_arr, c='r', label = 'Numerical')
plt.plot(t_arr, y_arr_ex, c='g', label = 'Analytical')
# plot properties
plt.xlabel('time')
plt.ylabel('position')
#plt.xlim(-0.5,7)
#plt.ylim(-5,200)
plt.legend()
plt.show()
```

```python
# plot output
plt.plot(t_arr, v_arr, c='b', label = 'Numerical')
# plot properties
plt.xlabel('time')
plt.ylabel('position')
#plt.xlim(-0.5,7)
#plt.ylim(-5,200)
plt.legend()
plt.show()
```

```python
t_arr_1 = []
t_arr_2 = []
Analitik_arr = []
Numerik_arr = []

y = 200
dt = 0.1
v = 0
a = 9.8
time = float(math.sqrt(2*y/a))

a = a * -1
yt_analitik = y

#NUMERIK   
vt_numerik = 0
yt_numerik = y
for x in np.arange(0, time+dt, dt):
# v(t) = v(t-1) + (a x ∆t)
#y(t) = y(t-1) + (v(t) x ∆t)
  vt_numerik = vt_numerik + (a * dt)
  yt_numerik = yt_numerik + (vt_numerik * dt)
  Numerik_arr.append(yt_numerik)
  t_arr_1.append(x)
  if yt_numerik < 0:
    print()
        
#ANALITIK    
for t in np.arange(0,time+dt,dt):
# exact solution ⟹ y(t) = y(0) + (1/2gt^2)
# v(t) = v(0) + g*t
  yt_analitik = y + (0.5 * a * t * t)
  t_arr_2.append(t)
  Analitik_arr.append(yt_analitik)
  vt_analitik = v + a * t

print("Numerik (yt) :", yt_numerik)
print("Numerik (vt) :", vt_numerik)
print("Analitik (yt) :", yt_analitik)
print("Analitik (vt) :", vt_analitik)
plt.xlabel('Time')
plt.ylabel('Height')
plt.plot(t_arr_1, Analitik_arr, 'r')
plt.plot(t_arr_2, Numerik_arr,'g')
plt.legend(['Analitik','Numerik'], loc = 'upper right')
plt.ylim(0,y)
plt.show()
```
